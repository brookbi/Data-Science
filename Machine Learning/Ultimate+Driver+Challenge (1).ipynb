```python
with open('C:/Testing/ultimate_data_challenge.json') as f:
    data = json.load(f)
```

```python
users = pd.DataFrame(data, columns=['avg_dist', 'avg_rating_by_driver', 'avg_rating_of_driver', 'avg_surge', 'city', 'last_trip_date', 'phone', 'signup_date', 'surge_pct', 'trips_in_first_30_days', 'ultimate_black_user', 'weekday_pct'])
```

```python
users
```

```python
#Diagnose missing data and incorrect data types.
users.info()
```

```python
#Trim off Rows with Empty Columns
users_clean = users[users['avg_rating_of_driver'].notnull()]
users_clean = users_clean[users_clean['avg_rating_by_driver'].notnull()]
users_clean = users_clean[users_clean['phone'].notnull()]
```

```python
users_clean.info()
```

```python
#Convert date columns into datetime data type
users_clean['last_trip_date'] = users_clean['last_trip_date'].apply(dateutil.parser.parse, yearfirst=True, dayfirst=False)
users_clean['signup_date'] = users_clean['signup_date'].apply(dateutil.parser.parse, yearfirst=True, dayfirst=False)
```

```python
users_clean['avg_dist'].mean()
```

```python
users_clean['avg_dist'].median()
```

```python
users_clean[users_clean['avg_dist'] <= 0.00]
#How do these 73 rows have avg distances of 0.00 miles?? Looking at the data and what qualifies for the avg_dist column,
#They shouldn't be here. Eliminating rows where a first trip's distance was supposedly '0' miles.
```

```python
users_clean = users_clean[users_clean['avg_dist'] != 0.0] 
```

```python
users_clean.iloc[0, 5] - users_clean.iloc[0, 7] <= pd._libs.tslibs.timedeltas.Timedelta('180 days 00:00:00')
```

```python
#Iterate to create a column of time deltas for how long it's been since a ride was taken. Takes a long time but works

idx = 0 
users_clean['days_since_last_ride'] = 0
while idx < len(users_clean):
    
    users_clean.iloc[idx, 12] = pd._libs.tslibs.timedeltas.Timedelta('180 days 00:00:00') - (users_clean.iloc[idx, 5] - users_clean.iloc[idx, 7])
    idx += 1
```

```python
idx = 0 
users_clean['retained'] = 0

while idx < len(users_clean) :
    if users_clean.iloc[idx, 12] <= pd._libs.tslibs.timedeltas.Timedelta('31 days 00:00:00') :
        users_clean.iloc[idx,13] = True
    else :
        users_clean.iloc[idx,13] = False
    idx += 1
```

```python
from statsmodels.formula.api import ols
```

```python
subriders = users_clean.loc[:, ('avg_dist', 'avg_rating_by_driver', 'avg_rating_of_driver', 'avg_surge', 'city', 'phone', 'surge_pct', 'trips_in_first_30_days', 'ultimate_black_user', 'weekday_pct', 'retained')]
```

```python
subriders['retained'] = subriders['retained'].astype(int)
len(subriders)
```

```python
subtrain = subriders[:24000]
subvalidate = subriders[24000:32000]
subtest = subriders[32000:]
```

```python
m = ols('retained ~ avg_dist + avg_rating_by_driver + avg_rating_of_driver + avg_surge + city + phone + surge_pct + trips_in_first_30_days + ultimate_black_user + weekday_pct', subtrain).fit()
print(m.summary())
```

```python
# Remove insignificant variables and repeat.

m = ols('retained ~ avg_dist + avg_rating_by_driver + city + phone + surge_pct + trips_in_first_30_days + ultimate_black_user + weekday_pct', subtrain).fit()
print(m.summary())
```

```python
# Remove insignificant variables and repeat.

m = ols('retained ~ avg_dist + avg_rating_by_driver + city + phone + surge_pct + trips_in_first_30_days + ultimate_black_user', subtrain).fit()
print(m.summary())
```

```python
# Training is complete. Moving on to Validation

import sklearn.metrics

m = ols('retained ~ avg_dist + avg_rating_by_driver + city + phone + surge_pct + trips_in_first_30_days + ultimate_black_user', subvalidate).fit()
print(m.summary())

preds = m.predict(subvalidate)
sklearn.metrics.mean_squared_error(subvalidate['retained'], preds)
```

```python
#No matter which variable is removed, the model gets worse. Thus, the current 
#Model with its significant variables is validated.

#This suggests that average distance of drives, avg rating by driver, city of origin,
#type of phone, percentage of rides taken during surge, and number of trips taken in a rider's first 30 days.
```

```python
preds = m.predict(subtest)
```

```python
preds_round = pd.Series.round(preds, 0)
```

```python
sklearn.metrics.accuracy_score(subtest['retained'], preds_round)
#Our Model correctly predicts whether an individual will be retained with 72% accuracy. 
```

```python
# I chose to use multiple linear regression with this model, given the fact that the 
# variable we're trying to predict is binary and we were given a medium-sized
# dataset to work with (~50k samples). The data size was large enough to avoid
# overfitting when splitting into train-validate-test sets. While categorization and classification
# were alternatives that were considered, ultimately this proved a quick and 
# straightforward approach to begin drawing insights from the data and analyze
# the predictive power of each individual variable in the set.

# All variables kept in the model are significant at a (P < 0.001) level, and 
#the overall accuracy score on the test set is roughly 72%. So while the model
#is not perfect, it does much better than chance alone. Given more data and 
#variables to experiment with, this score could continue to increase. 

# THe biggest insight to leverage are the variables that predict user retention. 
# First of all, Ultimate can work to reduce factors that might be turning 
# potential stable customers away from the application, such as excessive 
# surge prices or promote incentives for using the application frequently during
# the first 30 days. They could advertise more in certain cities than others,
# a more effective use of a marketing budget. Improving these factors
# could greatly influence rider retention.
```
