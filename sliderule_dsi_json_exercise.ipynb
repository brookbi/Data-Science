## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

```python
#So this one is definitely a bit of a workaround. I couldn't figure out how to group by the values inside of the list in the 
#dataframe column, so instead I iterated through the list (checking if there was more than one theme) and built a dictionary
#to count how many times each code appeared. 


themes = json_data.mjtheme_namecode
dict = {}
for theme in themes :
    
        
    if (theme[0]['code']) not in dict.keys() :
        dict[theme[0]['code']] = 1
    
    if (theme[0]['code']) in dict.keys() :
        dict[theme[0]['code']] += 1
    
    if len(theme) > 1 :
        
        if (theme[1]['code']) not in dict.keys() :
            dict[theme[1]['code']] = 1
    
        if (theme[1]['code']) in dict.keys() :
            dict[theme[1]['code']] += 1
        
    if len(theme) > 2 :
        
        if (theme[2]['code']) not in dict.keys() :
            dict[theme[2]['code']] = 1
    
        if (theme[2]['code']) in dict.keys() :
            dict[theme[2]['code']] += 1
    
    if len(theme) > 3 :
        
        if (theme[3]['code']) not in dict.keys() :
            dict[theme[3]['code']] = 1
    
        if (theme[3]['code']) in dict.keys() :
            dict[theme[3]['code']] += 1
            
print(dict)
```
